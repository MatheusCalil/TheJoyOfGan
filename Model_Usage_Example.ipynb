```python
from tensorflow import keras
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load the generator
last_generator = keras.models.load_model("/Users/matheus.faleiros/Documents/GitHub/TheJoyOfGan/Final/Model")
```

```python
#create the noise with bath_size, noise_size. Noise_size must be 100 because of input shape from the model
noise = tf.random.normal([15, 100])
```

```python
gen_image = last_generator(noise, training = False) #Generate 15 images
```

```python
for i in range(15): #display 15 images using matplotlib
    image = np.array(gen_image[i, :, :, :]* 127.5 + 127.5, dtype='uint8').reshape((1,128,128,3)) #Generated image are normalized
    plt.figure()
    plt.imshow(image[0])
```
